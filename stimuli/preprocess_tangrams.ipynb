```python
import numpy as np
import random

import glob
import os
from os import listdir
from os.path import isfile, join
from cairosvg import svg2png
from PIL import Image
import itertools
```

The goal of this notebook is to take a set of tangram SVGs from kilogram and convert them to PNGs without outlines in black. 

```python
mypath = "./raw_tangrams"
onlyfiles = glob.glob(mypath + '/*.svg', recursive=True)

for file in onlyfiles:
    print(file)
    relative_path = file[file.find('page'):]
    
    if 'page' in relative_path:
        print(relative_path)
        f = open("./raw_tangrams/" + relative_path,'r').read().replace("\n"," ")
        f = f.replace('fill="lightgray"','fill="black"')
    # f = f.replace('stroke="white" strokewidth="1"','stroke="black" strokewidth="2"')
        f = f.replace('stroke="white" strokewidth="1"','stroke="black" strokewidth="1"')
        svg2png(bytestring=f,write_to='./processed_tangrams/square-black-border/png/' + os.path.splitext(relative_path)[0]+'.png')
        with open("./processed_tangrams/square-black-border/svg/"+relative_path, 'w') as new_f:
            new_f.write(f)
```

```python
# make shape combos - diagonal

top_shape_pngs = sorted(glob.glob("./processed_tangrams/square-black-border/png/top/" + '/*.png', recursive=True))
bottom_shape_pngs = sorted(glob.glob("./processed_tangrams/square-black-border/png/bottom" + '/*.png', recursive=True))

top_imgs = [ Image.open(i) for i in  top_shape_pngs ]
bottom_imgs = [ Image.open(i) for i in  bottom_shape_pngs ]

for i in range(16):
    images = [top_imgs[i].rotate(180), bottom_imgs[i]]
    #min_shape = sorted( [(np.sum(i.size), i.size ) for i in images])[0][1]
    imgs_comb = np.vstack([i.resize((256, int(256/2))) for i in images])
    imgs_comb = Image.fromarray( imgs_comb)
    imgs_comb.save( './processed_tangrams/square-black-border/compositional/' + str(i) + "_" + str(i+16) + '.png' )   
```

```python
# make shape combos - small set of 16

for i in range(4):
    for j in range(4):
        images = [top_imgs[i].rotate(180), bottom_imgs[j]]
    #min_shape = sorted( [(np.sum(i.size), i.size ) for i in images])[0][1]
        imgs_comb = np.vstack([i.resize((256, int(256/2))) for i in images])
        imgs_comb = Image.fromarray( imgs_comb)
        imgs_comb.save( './processed_tangrams/square-black-border/compositional/' + str(i) + "_" + str(16+j) + '.png' )   
```
