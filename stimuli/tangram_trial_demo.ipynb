```python
def genNonComTrials(target_list, tops_list, bottoms_list, context_size = 9):
    all_trials = []
    tangrams_set = list(map(list, zip(tops_list, bottoms_list)))
    for target in target_list:
        filtered_tangrams_set = []
        for tangram in tangrams_set:
            if (tangram[0] != target[0]) & (tangram[1] != target[1]):
                filtered_tangrams_set.append(tangram)
        
        random.shuffle(filtered_tangrams_set)
        all_trials.append([target] + filtered_tangrams_set[0:context_size-1])
    return(all_trials)

print(genNonComTrials(non_comp_taget_list, tops_list, bottoms_list, context_size=9))
```

```python
def genBetweenCompTangrams(target_list, tops_list, bottoms_list, context_size = 9):
    all_trials = []
    for target in target_list:
        trial_list = []
        trial_list.append(target)
        #don't repeat the top
        filtered_tops_list = [x for x in tops_list if x != target[0]]
        new_tops_list = sample(filtered_tops_list[0:9], context_size-1)
        # dont repeat the bottom

        filtered_bottoms_list = [x for x in tops_list if x != target[1]]
        new_bottoms_list = sample(filtered_bottoms_list[0:9], context_size-1)
        
        for i in range(len(new_tops_list)):
            trial_list.append([new_tops_list[i], new_bottoms_list[i]])
        all_trials.append(trial_list)
    return(all_trials)
```

This notebook generates sets of trials based on the structure type and the # of shapes in-context.

Trial block, I need to ensure that there are a set of shapes that reoccurr as target shapes, so that I can measure conventions. I also need to make sure that the shapes chosen for the contexts fit the structure type. 

The 3 context structures are:
1) No compositionality - Each shape is uniquely and consistantly paired with one other shape. The stays throughout the experiment.
2) Within compositionality - Both shapes in the target shape appear as a distractor at least once. (e.g., AB, CD, AD, CB)
3) Between compositionality - Both shapes from the target shape appear just once, but can be reshuffled on following trials.
